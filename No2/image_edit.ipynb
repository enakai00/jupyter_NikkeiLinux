**[4-1]** 画像を操作するためのモジュールをインポートします。

```python
from PIL import Image
from PIL import ImageDraw
```

**[4-2]** インターネットから画像ファイル photo.jpg をダウンロードします。

```python
%%bash
curl -sOL https://raw.githubusercontent.com/enakai00/jupyter_ml4se/master/photo.jpg
ls -l photo.jpg
```

**[4-3]** 画像ファイルを読み込んで、変数 image に格納します。

```python
image = Image.open('photo.jpg')
image
```

**[4-4]** 変数 image に読み込んだ画像に線を書き加えます。

```python
draw = ImageDraw.Draw(image)
draw.line([(60,100),(180,100),(180,220),(60,220),(60,100)],
          width=5, fill='blue')
image
```
